```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv(r"YOUR_LOCATION")
dataset
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error as RMSE
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
x_train = dataset[['OptionType','Strike','Spot','TimeToExpiry','RiskfreeRate','MarketFearIndex','BuySellRatio']]
x_train = x_train.replace('Put',0)
x_train = x_train.replace('Call',1)
x_train
```

```python
y_train = dataset[['OptionPrice']]
y_train
```

```python

for i in range(1,15):
    poly = PolynomialFeatures(degree = i)
    x_poly = poly.fit_transform(x_train)
    lin_model = LinearRegression()
    lin_model.fit(x_poly,y_train)
    y_pred = lin_model.predict(x_poly)
    r2 = r2_score(y_train,y_pred)
    mse = RMSE(y_train,y_pred)
    print(f"The r2 score for {i} is {r2}")
    print(f"The MSE for {i} is {mse}")
```

```python
from sklearn.linear_model import Ridge
  # Try smaller degrees
poly = PolynomialFeatures(degree=7)
x_poly = poly.fit_transform(x_train)
    
ridge_model = Ridge(alpha=0.1)  # Add regularization
ridge_model.fit(x_poly, y_train)
y_pred = ridge_model.predict(x_poly)
    
r2 = r2_score(y_train, y_pred)
mse = RMSE(y_train, y_pred)
print(f"Degree {7}, RÂ²: {r2}, RMSE: {np.sqrt(mse)}")
```
